```python
import wikipedia as wiki
ngaag=wiki.summary("india")
ngaag
```

```python
import nltk
tokenizer=nltk.tokenize.WhitespaceTokenizer()
gaga=tokenizer.tokenize(ngaag)
```

```python
aa=nltk.pos_tag(gaga)
aa
```

```python
for i in range(0,len(aa)):
    if aa[i][1]=='NNP':
        print(aa[i][0])
```
